```python
import smtplib  
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

def send_feedback(comment_txt, comment_html, name='Neil Lawrence', user_id='ac1ndl', sender='N.Lawrence@sheffield.ac.uk'):
    # Credentials (if needed)  
    
    username = sender
    

    fromaddr = 'Neil Lawrence ' + '<'+sender+'>'  
    toaddr  = name + '<' + user_id + '@sheffield.ac.uk>'  

    cc = ['N.Lawrence@sheffield.ac.uk', 'm.zwiessele@sheffield.ac.uk', 'andreas.damianou@sheffield.ac.uk', 'acq11ra@sheffield.ac.uk']
    bcc =[]


    message_subject = "Assignment One Marks and Feedback"


    # Create message container - the correct MIME type is multipart/alternative.
    msg = MIMEMultipart('alternative')
    msg['Subject'] = message_subject
    msg['From'] = fromaddr
    msg['To'] = toaddr
    msg['CC'] = ', '.join(cc)

    # Create the body of the message (a plain-text and an HTML version).
    text = comment_txt
    html = """\
<html>
  <head><script type="text/x-mathjax-config">
MathJax.Hub.Config({
  tex2jax: {inlineMath: [['$','$'], ['\\(','\\)']]}
});
</script>
<script type="text/javascript"
  src="http://cdn.mathjax.org/mathjax/latest/MathJax.js?config=TeX-AMS-MML_HTMLorMML">
</script></script></head>
  <body>
     """ + comment_html + """
  </body>
</html>
"""

    # Record the MIME types of both parts - text/plain and text/html.
    part1 = MIMEText(text, 'plain')
    part2 = MIMEText(html, 'html')

    # Attach parts into message container.
    # According to RFC 2046, the last part of a multipart message, in this case
    # the HTML message, is best and preferred.
    msg.attach(part1)
    msg.attach(part2)

    toaddrs = [toaddr] + cc + bcc
    # The actual mail send  
    server = smtplib.SMTP('smtp.gmail.com:587')  
    server.starttls()  
    #server.login(username,password)  
    #server.sendmail(fromaddr, toaddrs, msg.as_string())  
    print msg.as_string()
    server.quit()
    print "Sent mail from", fromaddr, "to", toaddr


def process_answer(cell, c_type, part):
    """Process the cell, inserting comments and mark boxes at the appropriate places."""
    
    answer_marks = 'marks[' + str(assignment_info[part]['index']) + '] =' 
    total_marks = '# maximum marks for this part: ' + str(assignment_info[part]['marks'])
    
    new_cell = {u'cell_type': u'code', 
                u'language': u'python', 
                u'outputs': [], 
                u'collapsed': False, 
                u'input': [answer_marks, total_marks], u'metadata': {}}
    comment = {u'source': [u"<font color='red'>Comments on this answer " + part, assignment_info[part]['comment'] + ".</font>"], 
               u'cell_type': u'markdown', 
               u'metadata': {}}
    return new_cell, comment

def extract_question_part(string):
    """Extract the question part from the top line comment in the box."""
    try:
        found = re.search('Answer\s*([1-9]*[\(|\)|a-z]*\)).*', string).group(1).strip()
        
    except AttributeError:
        print "Couldn't find question part in ", string
        raise
    return found

def extract_comment_part(string):
    """Extract the comment part from the top line comment in the box."""
    try:
        found = re.search('answer\s*([0-9]*[\(|\)|a-z]*\)).*', string).group(1).strip()
    except AttributeError:
        print "Couldn't find question part in ", string
        raise
    return found

def extract_mark_index(string):
    """Extract the comment part from the top line comment in the box."""
    try:
        index = int(re.search('marks\[([0-9]*)\]', string).group(1))
    except AttributeError:
        print "Couldn't find index in ", string
        raise
    try:
        mark = float(re.search('marks\[[0-9]*\]\s*=\s*([0-9]*)', string).group(1))
    except AttributeError:
        print "Couldn't find index in ", string
        raise
    
    return index, mark
```

```python
# Search for the student file name.
def get_name(user_id):
    file_name = ''
    for file in os.listdir(submitted_path):
        if fnmatch.fnmatch(file, '*' + user_id + '*.txt'):
            file_name = file
    comments_str = open(os.path.join(submitted_path, file_name)).readline()
    string = comments_str
    try:
        name = re.search('Name:\s*(.*)\s*\(' + user_id + '\).*', string).group(1)
    except AttributeError:
        print "Couldn't find name in ", file_name
        raise
    return name    
```

```python

# Search for the student file name.
file_name = ''
for file in os.listdir(submitted_path):
    if fnmatch.fnmatch(file, '*' + user_id + '*.ipynb'):
        file_name = file
   
# Load in their ipython notebook file.
answer_str = open(os.path.join(submitted_path, file_name)).read()
answers = json.loads(answer_str)
answers_enhanced = answers.copy()
enhanced_cells = answers['worksheets'][0]['cells']

parts_used = []

# extact all the cells from the first worksheet (where the answers should be embedded)
cells = answers['worksheets'][0]['cells']
for cell in reversed(cells):
    if cell['cell_type']=='code':
        if len(cell['input'])>0 and len(cell['input'][0])>7 and cell['input'][0][:8] == '# Answer':
            part = extract_question_part(cell['input'][0])
            if part in parts_used:
                print "Error part ", part, " has already been introduced into the file."
                raise
            parts_used.append(part)
            new_cell, comment = process_answer(cell, c_type='code', part=part)
            i = cells.index(cell)
            cells.insert(i+1, comment)
            cells.insert(i+1, new_cell)
    elif cell['cell_type'] == 'markdown':
        if len(cell['source'])>0 and len(cell['source'][0])>7 and cell['source'][0][:8] == '**Answer':
            part = extract_question_part(cell['source'][0])
            if part in parts_used:
                print "Error part ", part, " has already been introduced into the file."
                raise
            parts_used.append(part)
            new_cell, comment = process_answer(cell, c_type='markdown', part=part)
            i = cells.index(cell)
            cells.insert(i+1, comment)
            cells.insert(i+1, new_cell)

for key in assignment_info.keys():
    if key not in parts_used:
        if assignment_info[key]['marks'] == 0 and assignment_info[key]['comment']=="":
            print "Ignoring missing dummy question part ", key
        else:
            print "Error: No question part ", key, " has been found in file ", file_name
            raise
file_name = os.path.join(submitted_path, 'marked', user_id + '.ipynb')
if os.path.exists(file_name):
    print "Error file ", file_name, " already exists. Delete and rerun."
    raise
with open(file_name, 'w') as outfile:
    json.dump(answers, outfile)
print "Finished ..."
```

```python
   
def post_process(file_name):  
    answer_str = open(os.path.join(submitted_path, 'marked', file_name)).read()
    answers = json.loads(answer_str)
    answer_cells = answers['worksheets'][0]['cells']
    index_used = []
    parts_used = []
    comments = {}
    marks = np.zeros(len(assignment_info.keys()))
    for cell in answer_cells:
        if cell['cell_type']=='code':
            if len(cell['input'])>0 and len(cell['input'][0])>7 and cell['input'][0][:6] == 'marks[':
                index, mark = extract_mark_index(cell['input'][0])
                if index in index_used:
                    print "Error index ", str(index), " has already got a mark associated."
                    raise
                index_used.append(index)
                marks[index] = mark
            
        elif cell['cell_type'] == 'markdown':
            if len(cell['source'])>0 and len(cell['source'][0])>7 and cell['source'][0][:42] == "<font color='red'>Comments on this answer ":
                part = extract_comment_part(cell['source'][0])
                if part in parts_used:
                    print "Error part ", part, " has already had a comment read."
                    raise
                parts_used.append(part)
                comment = ''
                for line in cell['source']:
                    comment += line
                comment = re.sub("<font color='red'>", '', comment)
                comment = re.sub('</font>', '', comment)
                comment = re.sub('Comments on this answer ', '', comment)
                comment = re.sub('\n', ' ', comment)
                comment = re.sub('\.\.\s*$', '.', comment)
                comment = re.sub('\!\.\s*$', '!', comment)
                comment = re.sub('\ +', ' ', comment)
                comment = re.sub('^\s*'+re.escape(part), '', comment)
                comment = re.sub('^\s*', '', comment)
                comment = re.sub('^[.?!:]*\s*', '', comment)
                
                comments[part]=comment
            
    for key in assignment_info.keys():
        if key not in parts_used:
            if assignment_info[key]['marks'] == 0 and assignment_info[key]['comment']=="":
                print "Ignoring missing dummy question part ", key
            else:
                print "Error: No question part ", key, " has been found in comments section. ", file_name
                raise
    return marks, comments

# postprocess
marks = {}
comments = {}
for file_name in sorted(os.listdir(os.path.join(submitted_path, 'marked'))):
    try:
        if file_name[-5:] == 'ipynb':
            user_name = file_name[:-6]
            marks[user_name], comments[user_name] = post_process(file_name)
                
    except:
        print "Error in ", file_name
        raise



grade_file = 'gc_COM4509.A.121061_column_2013-11-27-16-42-14.csv'
import csv
output_rows = []
with open(os.path.join(submitted_path, 'marked', grade_file), 'rb') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='"')

    output_rows.append(next(reader, None))  # print the headers
    
    for row in reader:
        user_id = row[2]
        if user_id in marks.keys():
            row[6] = str(marks[user_id].sum())
            comment_html = """We will do feedback sessions across Friday's lab session. If you haven't yet had a feedback session please select an appointment on <a href="http://tinyurl.com/o2kdrsz">this page</a>"""
            comment_html += '<h4>Total Mark: ' + str(marks[user_id].sum()) + '</h4>'
            
            comment_txt = """We will do feedback sessions across Friday's lab session. If you haven't yet had a feedback session please select an appointment using this link http://tinyurl.com/o2kdrsz"""
            comment_txt += '\n\nTotal Mark: ' + str(marks[user_id].sum())
            for key in sorted(comments[user_id].keys()):
                if assignment_info[key]['marks']>0:
                    comment_html += '<p><br><br><b>Question ' + key + '</b><br>'
                    comment_html += '<b>Comment:</b> '+comments[user_id][key]
                    comment_html += '<br>Mark: <b>' + str(marks[user_id][assignment_info[key]['index']]) + '</b> out of ' + str(float(assignment_info[key]['marks']))+'</p>'
                    comment_txt += 'Question ' + key + '\n'
                    comment_txt += comments[user_name][key] + '\n'
                    comment_txt += 'Mark: ' + str(marks[user_id][assignment_info[key]['index']]) + ' out of ' + str(float(assignment_info[key]['marks'])) + '\n\n'
            #print "Password needed here!"
            #raise
            password = ''
            if user_id == 'acp13lz':
            #user_id = 'ac1ndl'
            
                send_feedback(comment_txt, comment_html, get_name(user_id), user_id)#user_id)
                    
        else: 
            row[6] = '0.0'
            row[5] = 'No'
            
        output_rows.append(row)
with open(os.path.join(submitted_path, 'marked', 'marked_file.csv'), 'wb') as grade_file:
    writer = csv.writer(grade_file, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_ALL)
    for row in output_rows:
        writer.writerow(row)
```
